```python
sea.distplot(sns['Shipping_Cost'])
py.show()
```

```python
sea.distplot(sns['Sales'], hist=False)
py.show()
```

```python
sea.boxplot(sns['Order_Quantity'])
py.title('Order_Quantity')
py.show()
```

```python
sea.boxplot(x=sns['Order_Quantity'])
py.title('Order_Quantity')
py.show()
```

```python
sea.jointplot(x='Sales',y= 'Profit', data=sns)
py.show()
```

```python
sns =sns[(sns.Profit < 10000) & (sns.Sales < 20000)]

sea.jointplot(x='Sales',y='Profit',data= sns)
py.show()

```

```python
sns =sns[(sns.Profit < 100) & (sns.Profit > -100) & (sns.Sales < 200)]
sea.jointplot(x='Sales',y='Profit',data=sns,kind="hex",color="b")
py.show()
```
